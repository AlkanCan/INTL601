```python
import pandas as pd
```

```python
url="https://raw.githubusercontent.com/datasets/world-religion-projections/main/rounded_percentage.csv"
```

```python
df=pd.read_csv('by_rounded_percentage_share_csv.csv')
```

```python
df
```

```python
my_series = df.iloc[:,1:11].max(numeric_only=True, axis =1 )
my_series.name = "maxval"

# Include maxval in df
df['maxval']=my_series
```

```python
df
```

```python
df['Religion'] = df.loc[:,'Buddhists':'Unaffiliated'].eq(df['maxval'], axis=0).apply(lambda x: ','.join(df.columns[3:11][x==x.max()]),axis=1)
df
```

```python
df1=df[df['Year']==2010]
```

```python
df1.to_csv('religionall.csv')
```
